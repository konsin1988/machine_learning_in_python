### Decrease the dimensions

>Цель алгоритма уменьшения размерности - получить минимальный набор переменных, <br>
>который имитирует распределение переменных исходного набора данных. <br>
>Кроме того, сокращение числа переменных облегчает распознавание закономерностей, <br>
>включая естественные группировки, выбросы и аномалии. <br>

Важно отметить, что уменьшение размерности - это не просто удаление столбцов, а математическое преобразование информации в этих столбцах с целью получения информации с использованием меньшего количества переменных (столбцов). Например, если Вы изучаете цены на жилье, вы можете найти несколько коррелирующих переменных (например, площадь дома и почтовый индекс), которые можно объединить в новую переменную, адекватно отражающую эти две переменные. Таким образом, если применить сокращение размерности перед запуском основного алгоритма, модель будет работать быстрее, потреблять меньше вычислительных ресурсов и, возможно, обеспечит более точные прогнозы.

Еще одно преимущество этой техники - возможность визуализации многомерных данных. Помните, что максимальное количество измерений для диаграммы рассеяния - четыре. Хотя идеальными являются два или три измерения (четвертое измерение - время), алгоритмы уменьшения размерности могут быть использованы для упорядочивания набора данных с более чем четырьмя измерениями до четырех или менее переменных и проецирования синтетических переменных на визуальное пространство скаттерплота (графика рассеяния).

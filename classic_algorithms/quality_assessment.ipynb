# Показатели качества классификационной модели

## Accuracy
> Это простая метрика, измеряющая, сколько случаев модель классифицировала правильно, <br>
> деленная на полное количество случаев. Если все предсказания верны, показатель accuracy равен 1.0, <br>
> а если все случаи предсказаны неверно - 0.

## Матрица путаницы
>Матрица запутанности (также известная как матрица ошибок) - это простая таблица, <br>
>которая обобщает результаты работы модели, включая точное <br>
>количество ложноположительных и ложноотрицательных результатов.

## Precision
>Precision - это отношение правильно предсказанных истинно положительных случаев к общему<br>
>числу предсказанных положительных случаев. Высокий показатель точности означает<br>
>низкое количество ложных срабатываний.

$$ precision = \frac {NumberTruePositives}{NumberPredictedPositiveCases} $$

Эта метрика отвечает на вопрос о том, насколько точно модель предсказывает положительный результат. Другими словами, это способность модели не выдать отрицательный случай за положительный, что важно, например, в случае с тестами на наркотики.

## Recall

>Показатель recall модели аналогичен показателю precision, но в данном случае <br>
>представляет собой отношение правильно предсказанных истинных положительных <br>
>результатов к реальным положительным случаям. Другими словами, отзыв отвечает <br>
>на вопрос о том, сколько положительных результатов было правильно классифицировано<br>
>как положительные. Это можно понимать как способность модели идентифицировать<br>
>все положительные случаи. Обратите внимание, что числитель (вверху) одинаков<br>
>для precision и recall, а знаменатели (внизу) различны.

$$ recall = \frac {NumberTruePositives}{NumberActualPositiveCases} $$

## F1-score

>F1-score - это средневзвешенное значение precision и recall. Обычно он используется в качестве метрики для сравнения моделей между собой, а не для оценки точности отдельной модели.

>Кроме того, показатель f1-score обычно ниже, чем показатель точности, из-за способа расчета recall и precision.

# Показатели качества регрессионной модели

### MAE (mean absolute error) 
> измеряет средние ошибки в наборе прогнозов, то есть насколько далеко линия регрессии находится от реальных точек данных.

$$ MAE = \frac{1}{n}\sum_{n = 1}^{n} |y_i - x_i|$$

$$ y_i - predicted\ value\\ x_i - actual\ value \\ n - number\ of\ observations$$

### MSE (mean square error) 

>измеряет стандартное отклонение ошибок прогнозирования, которое показывает, насколько сконцентрированы или разбросаны ошибки прогнозирования от оптимального соответствия.

$$ MSE = \frac{1}{n}\sum_{i=1}^n (y_i-x_i)^2$$

$$ y_i - predicted\ value\\ x_i - actual\ value \\ n - number\ of\ observations$$

### MAPE (mean absolute percent error)

>Средний абсолютный процент ошибки измеряет средний процент отклонения предсказанных значений от реальных: MAPE особенно полезна для данных, где важна относительная ошибка, а не абсолютная величина ошибки 

$$ MAE = \frac{1}{n}\sum_{n = 1}^{n} \left\vert\frac{y_i - x_i}{y_i}\right\vert$$

$$ y_i - predicted\ value\\ x_i - actual\ value \\ n - number\ of\ observations$$

### RMSE (root mean )

>RMSE — это корень из MSE, который помогает вернуть ошибку в тех же единицах измерения, что и исходные данные. RMSE полезна в тех случаях, когда требуется более интерпретируемая оценка, чем просто MSE, поскольку её масштаб совпадает с масштабом целевой переменной.

$$ RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^n (y_i-x_i)^2}$$

$$ y_i - predicted\ value\\ x_i - actual\ value \\ n - number\ of\ observations$$

### $R^2$ — Коэффициент детерминации

>Еще одной важной метрикой для оценки качества линейной регрессии является коэффициент детерминации $R^2$, который показывает, какую долю изменчивости целевой переменной объясняет модель. Значение ближе к 1 указывает на лучшее качество модели — чем выше $R^2$, тем лучше модель объясняет данные.

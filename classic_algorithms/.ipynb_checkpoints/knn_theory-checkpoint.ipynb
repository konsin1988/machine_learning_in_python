# K-Nearest Neighbors

### 1. Евклидово расстояние
> In multi-dimensional space:

$$ d(A, B) = \sum_{i=1}^n (x_{i2} - x_{i1})^2 $$

**Преимущества:**

-    Легко интерпретировать.
-    Подходит для большинства задач, когда признаки находятся в одном масштабе.

**Недостатки:**

-    Чувствительна к разным масштабам признаков и выбросам.


### 2. Манхэттенское расстояние
>For two point:

$$ d(A, B) = |x_2 - x_1| + |y_2 - y_1|$$

**Преимущества:**

-    Удобно для данных с разной важностью отдельных признаков.
-    Менее чувствительна к выбросам по сравнению с Евклидовым расстоянием.

**Недостатки:**

-    Может не подходить для задач, где важны "напрямую" измеренные расстояния.

Метод k-nearest neighbor иногда называют «процедурой, основанной на памяти», поскольку при каждом прогнозировании используются все обучающие данные. По этой причине k-NN обычно не рекомендуется использовать для анализа больших наборов данных и измерения множества расстояний в высокоразмерных данных. Однако сокращение числа измерений с помощью алгоритма понижения размерности, такого как анализ главных компонент (PCA), или объединение переменных - это распространенная стратегия упрощения и подготовки k-NN-анализа.

### Реализация k-NN включает в себя следующие шаги:

- Импорт библиотек
-    Импорт набора данных
-    Удаление переменных
-    Масштабирование данных
-    Установка значений X и y
-    Применение алгоритма
-    Оценка предсказания
-    Оптимизация
-    Прогнозирование
